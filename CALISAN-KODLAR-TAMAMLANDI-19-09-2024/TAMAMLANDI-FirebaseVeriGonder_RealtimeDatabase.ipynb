```python
import base64
from PIL import Image

def base64_to_image(base64_string, output_file='output.png'):
  img_data = base64.b64decode(base64_string)
  with open(output_file, 'wb') as f:
    f.write(img_data)
  img = Image.open(output_file)
  img.show()


```
